## Revenue Distribution Analysis (Histogram)

### Step 1: Import Visualization Library

We use Matplotlib to create visualizations.
The pyplot module helps us draw charts easily.
We give it the alias `plt` to make the code shorter and cleaner.

### Step 2: Create a Figure

Before plotting, we create a blank canvas for our graph.
The `figsize` parameter controls the width and height of the chart.
Here, (8,5) means width = 8 units and height = 5 units.

### Step 3: Plot the Histogram

A histogram shows how values are distributed across different ranges.

- `df["Total Revenue"]` selects the revenue column.
- `bins=10` divides the revenue values into 10 ranges (or buckets).
- The height of each bar represents how many values fall into that range.

This helps us understand how revenue values are spread across transactions.

### Step 4: Add Title and Axis Labels

Adding labels makes the chart easier to understand.

- Title explains what the chart represents.
- X-axis shows the revenue ranges.
- Y-axis shows the frequency (how many transactions fall in each range).

 Step 5: Display the Chart

The `plt.show()` function tells Python to render and display the chart.
Without this, the graph may not appear.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["Total Revenue"], bins=10, edgecolor="black")

plt.title("Distribution of Total Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Number of Transactions")

plt.tight_layout()
plt.show()
```

### Insight from Histogram

The revenue distribution is positively skewed.
Most transactions fall in lower revenue ranges, while a few high-value transactions extend toward the higher end.

This suggests that a small number of transactions contribute significantly to total revenue.

## Outlier Detection Using Boxplot

### Why Are We Using a Boxplot?

A boxplot helps us understand:
- How the data is spread
- Where the median lies
- Whether there are extreme values (outliers)

This is important because sometimes a few very large values can distort averages.



### Understanding the Code

We use the following steps in the code:

1. `plt.figure(figsize=(8,5))`
   - Creates a blank canvas for the plot.
   - `figsize` controls width and height of the chart.

2. `plt.boxplot(df["Total Revenue"])`
   - Selects the "Total Revenue" column.
   - Creates a boxplot using those values.
   - Automatically calculates:
     - Q1 (25th percentile)
     - Median (50%)
     - Q3 (75th percentile)
     - Whiskers (range without outliers)
     - Dots for extreme values (if any)

3. `plt.title("Boxplot of Total Revenue")`
   - Adds a title to the chart.

4. `plt.ylabel("Total Revenue")`
   - Labels the vertical axis.

5. `plt.show()`
   - Displays the final chart.


### What the Boxplot Shows Visually

- The box represents the middle 50% of revenue values.
- The line inside the box represents the median.
- The whiskers show the normal range of data.
- Any points outside the whiskers are potential outliers.

```python
# Creating a boxplot for Total Revenue

plt.figure(figsize=(8,5))

# Boxplot helps detect extreme values (outliers)
plt.boxplot(df["Total Revenue"])

# Adding title
plt.title("Boxplot of Total Revenue")

# Adding label
plt.ylabel("Total Revenue")

plt.tight_layout()
plt.show()
```

### Insight from Boxplot

The boxplot shows that most transactions fall within lower revenue ranges.

However, several high-value transactions appear as outliers above the upper whisker.

This confirms that revenue distribution is positively skewed, with a small number of transactions contributing significantly higher revenue.

## Correlation Analysis

### Why Are We Calculating Correlation?

Correlation helps us understand relationships between numerical variables.

For example:
- Does Total Revenue increase when Units Sold increases?
- Does higher Unit Price lead to higher Profit?
- Are Revenue and Profit strongly related?

Correlation values range between -1 and +1:

- +1 → Strong positive relationship
- -1 → Strong negative relationship
- 0 → No relationship

This helps identify patterns between variables.

```python
# Calculating correlation matrix for numerical columns

correlation_matrix = df.corr(numeric_only=True)

correlation_matrix
```

## Scatter Plot: Revenue vs Profit

To visually examine the relationship between Total Revenue and Total Profit,
we create a scatter plot.

If the points form an upward pattern, it confirms a strong positive relationship.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df["Total Revenue"], df["Total Profit"])

plt.title("Revenue vs Profit")
plt.xlabel("Total Revenue")
plt.ylabel("Total Profit")

plt.show()
```

## Interpretation: Revenue vs Profit

The scatter plot shows a clear upward trend.

This confirms a strong positive relationship between Total Revenue and Total Profit.

As revenue increases, profit generally increases as well.

However, the relationship is not perfectly linear, meaning profit is also influenced by other factors such as costs.
