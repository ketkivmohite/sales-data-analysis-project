# Sales Data Analysis Project

## Exploratory Data Analysis (EDA) on Global Sales Dataset

### Project Objective

The objective of this project is to perform Exploratory Data Analysis (EDA) on a global sales dataset to identify:

- Total revenue generated
- Revenue distribution by region
- Most profitable product categories
- Profit comparison between sales channels
- Key business insights to support decision-making

### Tools & Technologies Used

- Python
- Pandas
- Matplotlib
- Jupyter Notebook

```python
# importing pandas library for data manipulation and analysis
import pandas as pd
```

```python
# Loading the sales dataset into a pandas DataFrame
df = pd.read_csv("sales.csv")

# Displaying first 5 rows of the dataset
df.head()
```

```python
# Checking dataset structure, data types, and missing values
df.info()
```

```python
# Generating statistical summary for numerical columns
df.describe()
```

```python
# Calculating total revenue generated in the dataset
df["Total Revenue"].sum()
```

```python
# Grouping data by Region and calculating total revenue per region
df.groupby("Region")["Total Revenue"].sum()
```

## Revenue by Region - Insights

- Sub-Saharan Africa generates the highest total revenue.
- Europe is the second strongest performing region.
- North America contributes the least revenue among regions.

Most Profitable Item

```python
# Finding total profit generated by each product category
df.groupby("Item Type")["Total Profit"].sum().sort_values(ascending=False)
```

## Profitability by Item Type - Insights

- Cosmetics is the most profitable item type, generating approximately $14.56M in total profit.
- Household and Office Supplies also contribute significantly to overall profitability.
- Fruits generate the lowest total profit among all item types.

Is Online or Offline sales more Profitable ? 

```python
# Comparing profitability between Online and Offline sales channels
df.groupby("Sales Channel")["Total Profit"].sum()
```

## Profitability by Sales Channel - Insights

- Offline channel contributes approximately $24.92M in total profit.

  
- Online channel generates around $19.25M in profit.

```python
# Importing matplotlib for data visualization
import matplotlib.pyplot as plt
```

```python
# Creating a bar chart for total revenue by region
df.groupby("Region")["Total Revenue"].sum().plot(kind="bar",figsize=(10,5))

# Adding chart title and labels
plt.title("Total Revenue by Region")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)

# Displaying the plot
plt.show()
```

## Visualization Insight

- The bar chart clearly shows Sub-Saharan Africa as the top revenue-generating region.
- Europe follows as the second strongest region.
- North America contributes the least revenue among all regions.
- The company may consider investigating growth opportunities in North America.
